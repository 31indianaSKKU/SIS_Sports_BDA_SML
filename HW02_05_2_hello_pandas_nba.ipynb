2. dataframe slicing, indexing


2-1. loc 활용하기:This function is used for labels --> Straightforward

```python
# 데이터셋의 특정 컬럼(Name, Team)만 보기
```

```python
df.loc[:,["Name","Team"]]
```

```python
# 첫 5개의 선수들 이름과 포지션 확인하기
```

```python
df.loc[0:5,["Name","Position"]]
```

```python
# 10번째 선수부터 5명의 선수들의 나이, 키, 몸무게 확인하기
```

2-2. iloc 활용하기:iloc: Purely integer-location based indexing for selection by position



```python
## 첫 3명의 선수 데이터 확인하기
```

```python
df.iloc[:3]
```

```python
# 첫 3명의 선수들의 이름과 포지션 확인하기
```

```python
df.iloc[:3,[0,3]]
```

```python
# 데이터셋의 마지막 3명의 선수 데이터 확인하기
```

```python
df.iloc[-3:]
```

4. Groupby 활용하기

```python
# 포지션 별 통계 내기
```

```python
df.groupby(["Position"]).mean()
```

```python
# 팀별 총 연봉 계산하기
```

```python
df.groupby(["Team"]).sum()
```
