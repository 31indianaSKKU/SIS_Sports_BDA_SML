```python
import numpy as np
import cv2
import io
import skimage.io as io
import json
import matplotlib.pyplot as plt
from matplotlib.image import imread
```

```python
# 추출한 json 파일과 이미지 파일 불러오기
image_fn = './test_1.jpg'
pred_fn = './response.json'
```

```python
# 이미지 읽어오기
image = io.imread(image_fn)
plt.axis('off')
plt.imshow(image)
plt.show()
```

```python
# json 파일 로드
with open(pred_fn) as json_file:
    pred_anns = json.load(json_file)
```

```python
pred_anns
```

```python
dt_bbox = pred_anns['people'][0]['body'][0]['bbox']
```

```python
dt_keypoint = pred_anns['people'][0]['body'][0]['keypoints'] 
```

```python
x = dt_keypoint[1::3]
y = dt_keypoint[0::3]
confidence = dt_keypoint[2::3]
print("X : {}\nY : {}\nV :{}".format(x, y, confidence))
```

```python
fig=plt.figure(figsize=(8,8))
image = io.imread(image_fn)
cv2.rectangle(image, 
              (int(dt_bbox[0]), int(dt_bbox[1])),
              (int(dt_bbox[2]), int(dt_bbox[3])),
              (255,0,0), 5)
plt.plot(x, y, 'r*', markersize=6)


# 특정 keypoint 확인하기
idx = 1
plt.plot(x[idx], y[idx], 'g*', markersize=6)


plt.imshow(image)
plt.show()
```
